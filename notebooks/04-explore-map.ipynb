```python
import pandas as pd

from mlxtend.file_io import find_files
from collections import Counter
from matplotlib import pyplot as plt
import folium
```

```python
proc_scrape = '../data/processed/scrape/'
scrape_fs = find_files(substring='-gps', path=proc_scrape, recursive=True)
```

```python
test = pd.concat([pd.read_csv(x) for x in scrape_fs])
```

```python
def add_circle_to_map(map_obj, circle):
    """
    """
    popup_text = circle['username'] + ": " + circle['message']
    return folium.CircleMarker(
        location = [circle['latitude'], circle['longitude']],
        popup=popup_text,
        radius=5,
        fill_color='#3186cc',
        color='#6eb2ea'
    ).add_to(map_obj)
```

```python
web_map = folium.Map(location=[39.5, -98.35], zoom_start=3, tiles='Stamen Toner')
test.iloc[:500].apply(lambda x: add_circle_to_map(web_map, x), axis=1)
#map(add_circle_to_map(web_map), test.iloc[0:500])

web_map.save('../viz/3-28/new-map.html')
#web_map
```

**#Hashtags**

```python
def get_hashtag_counts(df):
    """
    """
    #create the columns we want from the csv
    tag_cols = ['hashtag{}'.format(x) for x in range(1,6)]
    # get all the values at once, ina single list (INCLUDES nulls)
    hashtags = test[tag_cols].dropna(subset=tag_cols, how='all').values.flatten()
    # Filter to non-null vals
    hashtags = hashtags[pd.notnull(hashtags)]
    # make every str lowercase so we accurately count
    lowered_hashes = [x.lower() for x in hashtags]
    # count
    counts = Counter(lowered_hashes)
    # transform into a dataframe, using a hashtag as a row
    df = pd.DataFrame.from_dict(counts, orient='index').reset_index()
    # rename to something readable
    df.rename(columns={'index':'hashtag', 0:'count'}, inplace=True)
    
    return df.sort(['count'], ascending=[0])
```

```python
get_hashtag_counts(test)
```
